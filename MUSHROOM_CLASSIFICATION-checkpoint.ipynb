TRAINING THE MODEL

```python
X = mushroom.drop('mushroom_class',axis=1)
X.head()
```

```python
y = mushroom['mushroom_class']
y.head()
```

TRAINING PART

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X, y, test_size = 0.20, random_state=None)
```

```python
import sklearn
from sklearn.model_selection import KFold,cross_val_score
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier,AdaBoostClassifier,GradientBoostingClassifier
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
import warnings
warnings.filterwarnings('ignore')
from sklearn.model_selection import GridSearchCV
```

```python
model =[LogisticRegression(),KNeighborsClassifier(),GaussianNB(),SVC(),DecisionTreeClassifier(),RandomForestClassifier(),AdaBoostClassifier(),GradientBoostingClassifier()]

for m in model:
    kfold = KFold(n_splits=10,random_state=None)
    CV_m = cross_val_score(m,X_train,y_train,cv=kfold,scoring='accuracy')
    print(m)
    print(CV_m.mean())
    print('\n')
```

All the algorithms are showing excellent performance, some may be overfitting the model aswell.
Let's see which of these models predict the results preciesly without training them.

```python
for m in model:
    m.fit(X_train,y_train)
    pred = m.predict(X_test)
    print(m)
    print('Accuracy score = ',accuracy_score(pred,y_test))
    print('\n')
```

Seems that our model is performing well.
But what if we tune the parameters of the ensembling algorithms and then check how our model performs.
As doing this will somehow lower the risk of overfitting the model as well as predict more preciesly.

So let's choose any one of the ensembling techniques.

```python
RFC = RandomForestClassifier()
param_grid = {'n_estimators':[100,200,300,400,500,600,700,800,900,1000],
              'max_features':['auto','sqrt','log2'],
              'max_depth':[5,6,7,8,9,10],
              'criterion':['gini','entropy']}
GCV_RFC = GridSearchCV(estimator=RFC,param_grid=param_grid,cv=5)
GCV_RFC.fit(X_train,y_train)
```

```python
print('\n Best parameters = ')
print(GCV_RFC.best_params_)
```

```python
print('Best Scores = ',GCV_RFC.best_score_)
```

  Checking the important features.  

```python
plt.figure(figsize=(10,5))
final_model = RandomForestClassifier(criterion='gini', max_depth=8, max_features='auto', n_estimators=100)
final_model.fit(X,y)
importance = pd.Series(final_model.feature_importances_, index=X.columns)
importance.plot(kind='bar' , color = 'pink')
plt.title('Important Features')
plt.show()
```

```python
import joblib
joblib.dump(final_model,'mushroom.pkl')
```
