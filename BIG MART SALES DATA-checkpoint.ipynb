DEALING WITH OUTLIERS

```python
#z-score calculation
from scipy import stats
z = np.abs(stats.zscore(train))
print(z)
```

```python
#saving the filtered data into new_train
new_train = train[(z<3).all(axis=1)]
```

```python
new_train.isnull().sum()
```

```python
new_train.shape
```

```python
train.shape
```

```python
#Checking for skewness
new_cols = [cols for cols in new_train.columns if new_train[cols].dtype=='float64']
new_train[new_cols].skew()
```

```python
#Checking for zero values in float64 columns
(new_train[new_cols].values==0).any()
```

```python
#Applying BoxCox to bring the values in the data into a normal distribution
new_train[new_cols] = new_train[new_cols].mask(new_train[new_cols].values==0,0.0000001)
from scipy.stats import boxcox
for b in new_cols:
    new_train[b] = boxcox(new_train[b])[0]
new_train[new_cols].skew()
```

```python
new_train[new_cols].isnull().sum()
```

Now that the skewness is reduced aswell as the outliers are also removed. Seems that our data is noe clean.

```python
X_new = new_train[['Item_Fat_Content','Item_MRP', 'Outlet_Establishment_Year', 'Outlet_Size', 'Outlet_Location_Type', 'Outlet_Type']]
X_new.head()
```

```python
X_new_train,X_new_test,y_train,y_val = train_test_split(X_new,y,test_size=0.20,random_state=0)
```

```python
xgb_new = XGBRegressor()
xgb_new.fit(X_new,y)
xgb_new.score(X_new,y)
```

```python
results = cross_val_score(xgb_new,X_new_train,y_train,cv=10)
print(results.mean())
```

We can see that our results have increased by 2%, so we'll now try doing hypertunning and see the results.

```python
#Tunning the parameters
from sklearn.model_selection import GridSearchCV
params = {'booster':['gbtree','gblinear','dart'],
          'max_depth':[5,10,15],
          'n_estimators':[100,200,300],
          'reg_alpha':[0,1,5],
          'reg_lambda':[0,1,5]}
xgb_cv = GridSearchCV(estimator=xgb_new, param_grid=params,cv=10)
xgb_cv.fit(X_new_train,y_train)
```

```python
print(xgb_cv.best_params_)
```

```python
print(xgb_cv.best_score_)
```

```python
final_model = XGBRegressor(booster='dart', max_depth=5, n_estimators=100, reg_alpha=0, reg_lambda=5)
final_model.fit(X_new_train,y_train)
```

```python
final_model.score(X_new_train,y_train)
```

```python
pred = final_model.predict(X_new_test)
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
print('error:')

print('Mean absolute error:',mean_absolute_error(y_val,pred))
print('Mean squared error:',mean_squared_error(y_val,pred))

print('Root Mean Squared Error:',np.sqrt(mean_absolute_error(y_val,pred)))
print('R-Squared Error',r2_score(y_val,pred))
```

```python
print("Predicted:", pred)
print("Actual", y_val)
```

```python
sns.distplot(y_val,label='original')
sns.distplot(pred,label='predicted')
plt.title("Comparison")
plt.legend()
plt.show()
```
