TRAINING THE MODEL

```python
drop = ['Country','Region','Happiness Rank','Standard Error','Happiness Score']
X = whr.drop(drop,axis=1)
X.head()
```

```python
y = whr['Happiness Score']
y.head()
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.20,random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X_train,y_train)
lr.score(X_train,y_train)
```

```python
lr.intercept_
```

```python
lr.coef_
```

```python
coef_features = pd.DataFrame(list(zip(X.columns, lr.coef_)), columns=['features', 'coefficients'])
coef_features
```

```python
pred=lr.predict(X_test)
print("Predicted Happiness Score:", pred)
print("Actual Happiness Score", y_test)
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score
print('error:')

print('Mean absolute error:',mean_absolute_error(y_test,pred))
print('Mean squared error:',mean_squared_error(y_test,pred))

print('Root Mean Squared Error:',np.sqrt(mean_absolute_error(y_test,pred)))
print('R-Squared Error',r2_score(y_test,pred))
```

Our model is performing really well because errors are very less.

Now let's see how our model works if we train it using only these 4 features.

```python
X_new = whr[['Economy (GDP per Capita)', 'Family','Health (Life Expectancy)', 'Freedom','Generosity', 'Dystopia Residual']]
X_new.head()
```

```python
y.head()
```

```python
X_new_train,X_new_test,y_train,y_test = train_test_split(X_new,y,test_size=0.20,random_state=0)
```

```python
lr_new = LinearRegression()
lr_new.fit(X_new,y)
lr_new.score(X_new_train,y_train)
```

```python
#Using cross_val_score we can check the fitting of the model, whether underfitting or overfitting
from sklearn.model_selection import cross_val_score
cvs = cross_val_score(lr_new,X,y,cv=5,scoring='r2')
cvs
```

After performimg several iterations it was seen that while decreasing the number of features the score decreased by a percentage at each step.
So, there is no need of removing any feature as each of the feature is useful in a sense and removing any feature isn't really helping with the score improvement.
