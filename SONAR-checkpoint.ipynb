DEALING WITH OUTLIERS

```python
from scipy import stats
z = np.abs(stats.zscore(sonar))
print(z)
```

```python
#saving the filtered data into new_sonar
new_sonar = sonar[(z<3).all(axis=1)]
```

TRAINING THE MODEL

```python
X = new_sonar.iloc[:,0:-1]
X.head()
```

```python
y = new_sonar.iloc[:,-1]
y.head()
```

TRAINING PART 

```python
import sklearn
from sklearn.model_selection import KFold,cross_val_score
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.svm import SVC
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier,AdaBoostClassifier,GradientBoostingClassifier
from sklearn.metrics import accuracy_score,confusion_matrix,classification_report
import warnings
warnings.filterwarnings('ignore')

model=[LogisticRegression(),KNeighborsClassifier(),GaussianNB(),SVC(),DecisionTreeClassifier(),RandomForestClassifier(),AdaBoostClassifier(),GradientBoostingClassifier()]

for m in model:
    kfold = KFold(n_splits=10,random_state=None)
    result_cv = cross_val_score(m,X,y,cv=kfold,scoring='accuracy')
    print(m)
    print(result_cv.mean())
    print('\n')
```

Since, we have a confusion between Logistic Regression and AdaBoostClassifier we'll check the LEARNING SCORES OF THE ALGORITHMS.

```python
lg = LogisticRegression()
lg.fit(X,y)
lg.score(X,y)
```

```python
ABC = AdaBoostClassifier()
ABC.fit(X,y)
ABC.score(X,y)
```

AdaBoostClassifier clearly shows overfitting of model.
So, lets try tunning the parameters.

```python
from sklearn.model_selection import GridSearchCV
abc = AdaBoostClassifier()
param_grid = {'n_estimators':[100,200,300,400,500,600,700,800,900,1000],
              'learning_rate':[0.001,0.01,0.1,0.2,0.5,1],
              'algorithm':['SAMME','SAMME.R']
             }
GCV = GridSearchCV(estimator=abc,param_grid=param_grid,cv=5)
GCV.fit(X,y)
```

```python
print('\n Best parameters = ')
print(GCV.best_params_)
```

Now we have the best parameters of the algorithm to use for our final model training!

```python
model_1 = AdaBoostClassifier(algorithm='SAMME.R', learning_rate=1, n_estimators=200)
model_1.fit(X,y)
print('Learning Score = ',model_1.score(X,y))
```

```python
pred = model_1.predict(X)
print(pred)
```

```python
print(accuracy_score(y,pred))
print(confusion_matrix(y,pred))
print(classification_report(y,pred))
print('\n')
```

Even after tunning it is showing 100% accuracy which may possibly overfit.

# So, the best algorithm for this dataset appears to be logistic regression.

```python
final_model = LogisticRegression()
final_model.fit(X,y)
print('Learning Score = ',final_model.score(X,y))
```

```python
prediction = final_model.predict(X)
print(prediction)
```
