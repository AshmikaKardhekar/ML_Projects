Total (States), Total (UTs), Total (All-India) are the columns which provide us the total information.
Let's store this data into a different dataframe.

```python
temp = crime.groupby(['STATE/UT','CRIME']).sum()
temp.drop(index=['Total (States)', 'Total (UTs)', 'Total (All-India)' ],inplace=True)
```

```python
temp
```

Checking the CRIME in terms of STATE/UT.

```python
crime_data = temp.groupby(['CRIME','STATE/UT']).sum()
crime_data
```

1. Immoral Traffic (P) Act :- 

```python
sns.set(style="whitegrid")
crime_data.loc['Immoral Traffic (P) Act',:].plot.bar(figsize=(20,5),stacked=True)
plt.legend(loc='best')
plt.xticks(rotation = 90)
```

2. Top 5 states are Andhra Pradesh, Tamil Nadu, Karnataka, Maharashtra, Kerala.
3. It seems that the rate of this crime is more in the southern part of India.
4. Also, no to least crime is in the nothern part of India.
5. There is no to least crime in Lakshadweep, Manipur, Sikkim, Tripura
